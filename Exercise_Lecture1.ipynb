```python
#Exercise_1.2 Convert from metres to British length units
import pybryt
import pathlib
import numbers
import numpy as np

def pybryt_reference(lecture, exercise):
    
    dirname = pathlib.Path("pybryt-references")
    basename = f"exercise-{lecture}_{exercise}"
    pyfilename = dirname / f"{basename}.py"
    pklfilename = dirname / f"{basename}.pkl"

    if pyfilename.is_file():
        pybryt.ReferenceImplementation.compile(str(pyfilename)).dump(str(pklfilename))
    elif not pklfilename.is_file():
        msg = f"PyBryt reference {pklfilename!r} not found."
        raise FileNotFoundError(msg)

    return str(pklfilename)

#Load the reference (robust to naming)
ref_path = "pybryt-references/exercise-1_2.pkl"
#ref_path = pybryt_reference(1,2)
print("Using reference:", ref_path)

with pybryt.check (ref_path):
    metres = 640
    inches = metres / 0.0254      # 1 inch = 0.0254 m
    feet   = metres / 0.3048      # 1 foot = 0.3048 m
    yards  = metres / 0.9144      # 1 yard = 0.9144 m
    miles  = metres / 1609.34     # 1 mile = 1609.34 m

# ---- Optional sanity checks & prints (outside the block) ----
assert all(isinstance(i, numbers.Real) for i in [feet, yards, miles])
assert np.isclose(feet, 2099.737532808399)
assert np.isclose(yards, 699.912510936133)
assert np.isclose(miles, 0.3976775630318938)
print("feet:", feet, "| yards:", yards, "| miles:", miles)
print("All sanity checks passed âœ…")
```
